First line gets a playlist object, which has a number of keys. One of those keys is "tracks", which contains an	array of playlist track objects inside a paging object. 

(The offset-based paging object is a container for a set of objects. It contains a key called items (whose value is an array of the requested objects) along with other keys like previous, next and limit that can be useful in future calls.)

Playlist track object contains among others track objects, where information like artist or song name can be found. The 3 lines of code below assign "a collection containing a link ( href ) to the Web API endpoint where full details of the playlist’s tracks can be retrieved, along with the total number of tracks in the playlist" to the variable "songs".

```python
playlistTwoId="5bVHqd8G0QOA3pmo3mLNdy"
playlistTwo = sp.user_playlist(cid, playlistTwoId)
P2Tracks = playlistTwo["tracks"]
P2Songs = P2Tracks["items"] 
```

To get the information about every song from a long (> 100 songs) playlist, one needs to iterate through the tracks object, because spotify uses paginaton in this case.

```python
while P2Tracks['next']:
    P2Tracks = sp.next(P2Tracks)
    for item in P2Tracks["items"]:
        P2Songs.append(item)
```

Get Spotify IDs for every song. 

```python
P2Ids = [] 
for i in range(0,len(P2Songs)):
    P2Ids.append(P2Songs[i]['track']['id'])
```

```python
print(P2Ids)
```

```python
features = []
for i in range(0,len(P2Ids),50):
    audio_features = sp.audio_features(P2Ids[i:i+50])
    for track in audio_features:
        features.append(track)
        features[-1]['target'] = 1
```

Audio features call can only download 50 features at once, hence the 50 number in the loop.

```python
print(features)
```

```python
trainingData = pd.DataFrame(features)
```

```python
P2SongNames = []
for i in range(0,len(P2Songs)):
    P2SongNames.append(P2Songs[i]['track']['name'])

P2SongArtists = []
for i in range(0,len(P2Songs)):
    P2SongArtists.append(P2Songs[i]['track']['artists'][0]['name'])
```

```python
trainingData["SongName"] = P2SongNames
trainingData["SongArtist"] = P2SongArtists
```
